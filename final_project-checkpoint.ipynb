```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#for the CNN model
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import backend as k
```

```python
#About the Data
#The dataset consist of 27,455 training images and 7,172 test images. Each image is grayscaled and has a 28 x 28 pixel structure. Each images includes the label of which letter in the alphabet the image represents. There are 24 letters in this dataset, 'j' and 'z' are not included because they both involve motion.
#https://www.kaggle.com/datasets/datamunge/sign-language-mnist
```

```python
#downloading the training data
trainset =pd.read_csv("C:/Users/user/Desktop/project/sign_mnist_train/sign_mnist_train.csv")
trainset.head()
```

```python
testset = pd.read_csv("C:/Users/user/Desktop/project/sign_mnist_test/sign_mnist_test.csv")
testset.head()

```

```python
#Checking for Missing Data
#summing the number of na in the training set for each column
print(sum(trainset.isna().sum()))

#summing the number of na in the test set for each column
print(sum(testset.isna().sum()))
```

```python
#summing the number of null values in the training set for each column
print(sum(trainset.isnull().sum()))

#summing the number of null values in the test set for each column
print(sum(testset.isnull().sum()))
```

```python
#Processing the Images
# Creating our X and our Y
#creating our X for the training data
X_trainset=trainset.iloc[:,1:]
#creating our Y for the training data
y_trainset=trainset.iloc[:,0]
#creating our X for the test data
X_testset=testset.iloc[:,1:]
#creating our y for the test data
y_testset=testset.iloc[:,0]
```

```python
#Scaling Our Images
#converting the range of the pixel data from 0-255 to 0-1
X_trainset = X_trainset / 255.0


X_testset = X_testset / 255.0
```

```python
#Reshaping Our Images
X_trainset = X_trainset.values.reshape(-1,28,28,1)
X_testset = X_testset.values.reshape(-1,28,28,1)
print(X_trainset.shape)
print(X_testset.shape)
```

```python
#Viewing the Images
#1) Training Images
#creating a 5x5 grid of the first 25 photos in the training images
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(15, 10),
                        subplot_kw={'xticks': [], 'yticks': []})

for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(X_trainset[i], cmap='gray')
    plt.title(y_trainset[i])
plt.show()
```

```python

#2)Test Images
#creating a 5x5 grid of the first 25 photos in the test images
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(15, 10),
                        subplot_kw={'xticks': [], 'yticks': []})

for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(X_testset[i], cmap='gray')
    plt.title(y_testset[i])
plt.show()
```

```python

#Spliting Our Training Data
#spliting training images into the images we will use for training the model and validating the model
X_trainset, X_val, y_trainset, Y_val = train_test_split(X_trainset, y_trainset, test_size = 0.2, random_state=0)

```

```python
#Creating Our Model
#creating our CNN model
model = keras.Sequential([
    
    layers.BatchNormalization(),
    layers.Conv2D(filters=32, kernel_size=(5,5), activation="relu", padding='same',
                  input_shape=[28, 28, 1]),
    layers.MaxPool2D(),
    layers.Dropout(.25),
    
    layers.BatchNormalization(),
    layers.Conv2D(filters=32, kernel_size=(3,3), activation="relu", padding='same'),
    layers.MaxPool2D(),
    layers.Dropout(.25),
    
    layers.BatchNormalization(),
    layers.Conv2D(filters=64, kernel_size=(3,3), activation="relu", padding='same'),
    layers.MaxPool2D(),
    layers.Dropout(.25),

    layers.BatchNormalization(),
    layers.Conv2D(filters=128, kernel_size=(3,3), activation="relu", padding='same'),
    layers.MaxPool2D(),
    layers.Dropout(.25),
    
    layers.Flatten(),
    layers.Dropout(.25),
    layers.Dense(units=64, activation="relu"),
    layers.Dense(units=26, activation="softmax"),
])
```

```python
#compiling the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(epsilon=0.01),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#Training the model
history = model.fit(
    x = X_trainset,
    y = y_trainset,
    validation_data= (X_val,Y_val),
    batch_size = 128,
    epochs=30,
    verbose=2,
)
```

```python
#Viewing the training results
history_frame = pd.DataFrame(history.history)

history_frame.loc[:, ['accuracy', 'val_accuracy']].plot();
```
